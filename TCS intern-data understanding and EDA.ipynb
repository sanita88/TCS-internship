An individual’s annual income results from various factors like individual’s education level, age, gender, occupation, and etc.
Here the prediction task is to determine whether a person makes over 50K a year.

#The dataset contains 16 columns<br>
#Target filed: Income<br>
The income is divide into two classes: <=50K and >50K<br>
#Number of attributes: 14<br>
These are the demographics and other features to describe a person<br>

We can explore the possibility in predicting income level based on the individual’s personal information.<br>

```python
##importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('adult.csv')  ## reading the datset to python environment
```

```python
## Replacing special characters with nan
data['native-country'] = data['native-country'].replace('?',np.nan)
data['workclass'] = data['workclass'].replace('?',np.nan)
data['occupation'] = data['occupation'].replace('?',np.nan)
```

```python
data.isin(['?']).sum(axis=0) 
```

# Relation between different numerical columns and target column(income)

```python
sns.catplot(data=data, x='income', y='age', kind= 'box')
plt.title('Relationship between Income and Age')
plt.show()
```

Age of people with income less than 50K is between 25 and 45<br>
Age of people with income more than 50K is between 35 and 50<br>
Salary increment is happening with age

Self-emp-inc workclass is getting the highest income.

```python
fig = plt.figure(figsize=(10,10)) 
plt.title('Relationship between working hous per week and income')
sns.boxplot(x="income", y="hours-per-week", data=data)
plt.show()
```

As working hours per week increases income also increases

# Relation between different catogorical columns and target column(income)

```python

for col in (catag_cols):
    sns.countplot(data[col], hue=data["income"])
    plt.xticks(rotation=90)
    plt.show()
```

-Workclasses : number of people earning less then 50k are more then those earning more than 50k.Data mostly consist of employees from private sector<br>
-Education:data mostly consist of people who has education as hs-grad.People with bachelors degree is earning more.<br>
-Marital status:The people with marital status as Married-civ-spouce has highest people with income more then 50k.<br>
-Occupation:Most of the people who have income more then 50k either have prof-speciality or exec-managerial as occupation<br>
-Relationship:Mostly a person with relation as husband in a family has most count of people with more then 50k income<br>
-Race and gender:Male white people is earning more<br>
-Native country:The data mainly consist of people from US.
